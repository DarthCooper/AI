```python
df_HighBPM = df[df["bpm"] >= 122.54039874081847]
df_LowBPM = df[df["bpm"] <= 122.54039874081847]

def overlayHistograms(bottom, top, x, bins, color = "blue"):
    sns.histplot(data = bottom, x = x, bins = bins, color = "orange")
    sns.histplot(data = top, x = x, bins = bins, color = color)
    
overlayHistograms(df_LowBPM, df_HighBPM, "danceability_%", 50)
KeyMethods.getMean(df_LowBPM, "danceability_%")
```
