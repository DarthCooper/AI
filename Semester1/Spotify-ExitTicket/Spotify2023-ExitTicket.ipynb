# Spotify 2023 Danceability (%) Distribution
### The Max Dancibility is 96 % with the song "Peru" by Ed Sheeran
### The 3rd Quartile is 78%
### The Median is 69%
### The 1st Quartile is 57%
### The Min Dancibility is 23% with the song "White Christams" by Bing Crosby

```python
sns.boxplot(data=df, x="danceability_%")
summary = {
    "max": df["danceability_%"].max(),
    "Q3" : df["danceability_%"].quantile(0.75),
    "Median" : df["danceability_%"].median(),
    "Q1" : df["danceability_%"].quantile(0.25),
    "Min" : df["danceability_%"].min(),
}
summary
```
